### Limpieza de datos y estandarización

```python
productos.dropna() # Eliminación de datos núlos en 'Productos'
```

```python
# Corrección de formato de fechas para valoración precisa
# NOTA: Se solicitó apoyo de un tutor para obtener esta fórmula.

def corregir_week(valor):

  try:
    if '-' in valor and len(valor.split('-')[0]) == 2:
      mes_o_semana, year = valor.split('-')
      mes_o_semana = int(mes_o_semana) # Convierte a entero para validación

      if 1 <= mes_o_semana <= 12:
        return f"0{mes_o_semana}-22" if mes_o_semana < 10 else f"{mes_o_semana}-22"
      else:
        return valor
    return valor
  except Exception as e:
    return valor

# Aplica la corrección a la columna WEEK

fechas['WEEK'] = fechas['WEEK'].apply(corregir_week)
fechas.to_csv('fechas_corr.csv', index=False)
print("¡Corrección completada!")
```

```python
#Comprobación de correción de formato de fechas

fechas_corr = pd.read_csv('fechas_corr.csv')

fechas_corr['WEEK'].unique()
```

### Uniones y creación de tablas

```python
ventas_productos = pd.merge(ventas, productos[['ITEM', 'MANUFACTURER', 'BRAND']], left_on='ITEM_CODE', right_on='ITEM',how='left')

ventas_productos.head()
```

```python
category_segment = pd.merge(categorias, segmentos[['ATTR1', 'FORMAT', 'SEGMENT', 'CATEGORY']], left_on='ID_CATEGORY', right_on='CATEGORY', how = 'left')

category_segment.head()
```

```python
fecha_productos = pd.merge(ventas_productos, fechas_corr, left_on='WEEK', right_on='WEEK', how='left')

fecha_productos.head()
```

```python
fvp = pd.merge(ventas_productos, fechas_corr, left_on='WEEK', right_on='WEEK', how='left')

fvp.head()
```

#### Agrupaciones

```python
# Ventas por región

region_sales = fvp.groupby(['REGION'])['TOTAL_VALUE_SALES'].sum().sort_values(ascending=False)
regiones = region_sales[0:7]

regiones
```

```python
# Ventas por producto

brand_sales = fvp.groupby(['BRAND'])['TOTAL_VALUE_SALES'].sum().sort_values(ascending=False)

brand_sales
```

```python
# Ventas por año

year_sales = fvp.groupby(['YEAR'])['TOTAL_VALUE_SALES'].sum().sort_values(ascending=False)

year_sales
```

```python
# Ventas por semana

week_sales = fvp.groupby(['WEEK'])['TOTAL_VALUE_SALES'].sum().sort_values(ascending=False)

week_sales
```

### DF_FINAL

```python
df_productos = productos[['MANUFACTURER', 'BRAND', 'ITEM', 'CATEGORY', 'FORMAT']]

df_productos.head()
```

```python
product_category = pd.merge(df_productos, categorias, left_on='CATEGORY', right_on='ID_CATEGORY')

product_category.head(15)
```

```python
product_sales = pd.merge(product_category, ventas, left_on='ITEM', right_on='ITEM_CODE')

product_sales.head()
```

```python
df_calendar = fechas[['WEEK', 'YEAR', 'MONTH']]

df_calendar.head(10)
```

```python
sales_date = pd.merge(product_sales, df_calendar, left_on='WEEK', right_on='WEEK')

sales_date.head()
```

```python
df_unido = pd.merge(sales_date, segmentos, left_on='CATEGORY_x', right_on='CATEGORY')

df_unido.head()
```

```python
df_final = df_unido[['WEEK', 'MONTH', 'YEAR', 'ID_CATEGORY', 'MANUFACTURER',  'BRAND', 'SEGMENT', 'FORMAT_x', 'ITEM', 'TOTAL_UNIT_SALES', 'TOTAL_VALUE_SALES', 'REGION']]
df_final.head(50)
```

```python
df_final2 = df_final.drop_duplicates(df_final.columns[df_final.columns.isin(['TOTAL_VALUE_SALES'])], keep='first')
df_final2.head(20)
```
