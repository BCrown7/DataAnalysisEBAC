#### *Obtener 5 Palabras Clave*

```python
# Exploración de palabras clave en los tweets
palabras_clave = ['quality', 'service', 'price', 'deliver', 'satisfaction']
for palabra in palabras_clave:
    coincidencias = df[df.apply(lambda row: row.astype(str).str.contains(palabra, case=False).any(), axis=1)]
    print(f'Palabra clave: {palabra} - Coincidencias: {len(coincidencias)}')
    display(coincidencias.head(3))
```

#### *Importar por lo menos 1000 tweets a la base de datos a una nueva estructura.*

```python
# Tweets Structure
# Se insertan a la estructura los primeros 1000 tweets

df_1000 = df.head(1000)
tweets_structure = db['tweets_structure']
result = tweets_structure.insert_many(df_1000.to_dict('records'))
print(f'Se importaron {len(result.inserted_ids)} tweets a la colección tweets_structure.')
```

#### *Construir 2 filtros de búsqueda que se considere importante dentro de la base de datos de tweets proporcionada*

```python
# Filtro 1: Buscar tweets que contengan una palabra clave específica en cualquier campo
palabra_clave = 'servicio'
resultados_palabra = tweets_structure.find({"$or": [ {k: {"$regex": palabra_clave, "$options": "i"} } for k in df_1000.columns ] })
print(f'Tweets que contienen la palabra "{palabra_clave}":')
for tweet in list(resultados_palabra)[:5]:
    print(tweet)

# Filtro 2: Buscar tweets por usuario (ajusta el nombre de la columna según tu archivo, por ejemplo 'usuario' o 'user')
usuario = 'usuario_ejemplo'
resultados_usuario = tweets_structure.find({"usuario": usuario})
print(f'Tweets publicados por el usuario "{usuario}":')
for tweet in list(resultados_usuario)[:5]:
    print(tweet)
```
