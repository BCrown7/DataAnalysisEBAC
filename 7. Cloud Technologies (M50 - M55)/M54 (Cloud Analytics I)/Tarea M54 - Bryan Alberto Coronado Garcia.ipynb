- Incluir KPIs y datos que permitan a una persona sin conocer el negocio a fondo, darse cuenta de sus magnitudes (e.g. promedio del tamaño de una casa).

```python
# ------------------------------------------------------------------------
# Cuáles son los 10 códigos postales con el promedio más alto en precios?
# ------------------------------------------------------------------------

psql.sqldf("SELECT zipcode, \
            AVG(price) as PrecioPromedio \
            FROM df \
            GROUP BY zipcode \
            ORDER BY price \
            limit 10")

# Esta consulta nos ayuda a comprender y valorar los zipcodes mas valiosos para posibles inversiones
# ante los clientes, asi como para la empresa, ya que estos datos pueden ser utilizados
# para la toma de decisiones estratégicas en el negocio inmobiliario.
# Adicionalmente, se puede observar que los precios promedio de las casas en estos códigos postales
# son significativamente más altos que en otros códigos postales, lo que indica una alta demanda
# y un mercado inmobiliario activo en estas áreas. Lo que nos puede ayudar a entender alguna posible
# centrificacion de la zona, o el crecimiento de la misma.
```

```python
# ------------------------------------------------------------------------------
# Cuánto aumenta el precio por cada piso en una casa?
# -------------------------------------------------------------------------------

psql.sqldf("SELECT floors, \
            AVG(price) as PrecioPromedio \
            FROM df \
            GROUP BY floors \
            ORDER BY floors")

# Esta consulta nos permite confirmar el aumento estimado del precio por cada piso en una casa.
# Podemos observar que el precio promedio de las casas aumenta a medida que aumenta el número de pisos
# en la casa. Esto indica que las casas con más pisos tienden a tener un precio más alto, lo que puede ser
# un factor importante a considerar al evaluar el valor de una propiedad.
# También puede ser útil para la empresa al evaluar el valor de las propiedades y establecer precios
# competitivos en el mercado inmobiliario.
```

```python
# ------------------------------------------------------------------------------
# Las casas más antiguas siguen siendo rentables?
# ------------------------------------------------------------------------------
psql.sqldf("SELECT yr_built as AñoDeConstruccion, \
            AVG(price) as PrecioPromedio, \
            price/yr_built as PrecioPorAñoDeConstruccion \
            FROM df \
            GROUP BY yr_built \
            ORDER BY yr_built ASC \
            limit 21")

# Esta consulta nos ayuda a entender la relación entre la antigüedad de las casas y su precio.
# Podemos verificar si las casas más antiguas tienden a tener precios más bajos,
# lo que podría indicar una menor demanda o una percepción de menor valor en el mercado.
# También, al observar el precio por año de construcción, podemos identificar si hay una tendencia
# a que las casas más antiguas tengan un precio por año de construcción más bajo,
# lo que podría indicar una menor rentabilidad en el mercado inmobiliario para estas propiedades, a su vez
# valorar alguna posible renovación o remodelación de las mismas para aumentar su valor.
```
